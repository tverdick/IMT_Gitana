```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
```

```python
#Import des données
df = pd.read_csv('final_data2.csv',sep=';')
df = pd.DataFrame(df)
```

```python
#Vérification du type de chaque variable
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df.dtypes)
```

```python
#Supression des variables avec trop de NaN
for x in df : 
    if df[x].isnull().sum(axis = 0) > 12000:
        df = df.drop([x],axis=1)
df = df.dropna()
```

```python
#Selection des variables actionnables par le skipper
df1 = df[['Board_Elevator_Angle_Center',
         'Board_Trimmer_Angle_Center',
         'Mainsail_Load',
         'Mast_Rotation_Angle',
         'Rudder_Elevator_Angle_Center',
         'Foil' ,
         'Foil_Rake',
         'Rudder_Elevator',
         'Boat_Speed'
        ]]
```

```python
# imports
import pandas as pd
import statsmodels.api as sm

# assigner la vitesse du bateau comme y et toutes les autres variables comme x
variables = list(df1.columns)
y = 'Boat_Speed'
x = [var for var in variables if var not in y ]
#y = 'A'
#x = [var for var in variables if var not in y ]

#Regression linéaire
model_Simple = sm.OLS(df1[y], df1[x]).fit()

model_Simple.summary()
```
